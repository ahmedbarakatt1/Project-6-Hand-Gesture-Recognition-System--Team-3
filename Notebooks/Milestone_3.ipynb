```python
import cv2
import numpy as np
import tensorflow as tf

# --------------------------
# CONFIG
# --------------------------
MODEL_PATH = "hand_gesture_model.keras"   # put .keras model in same folder
IMG_SIZE = 128

gesture_classes = [
    "01_palm",
    "02_l",
    "03_fist",
    "04_fist_moved",
    "05_thumb",
    "06_index",
    "07_ok",
    "08_palm_moved",
    "09_c",
    "10_down"
]

# --------------------------
# Load Model
# --------------------------
print("Loading model...")
model = tf.keras.models.load_model(MODEL_PATH)
print("Model loaded successfully!\n")

# --------------------------
# Webcam Capture
# --------------------------
cap = cv2.VideoCapture(0)

print("Press SPACE to capture an image.")
print("Press Q to quit.")

captured = None

while True:
    ret, frame = cap.read()
    if not ret:
        print("❌ Webcam not detected.")
        break

    cv2.imshow("Camera - Press SPACE to capture", frame)
    key = cv2.waitKey(1) & 0xFF

    if key == ord('q'):
        print("Exiting...")
        break

    if key == ord(' '):  # SPACE
        captured = frame.copy()
        cv2.imwrite("captured.jpg", captured)
        print("Image captured → saved as captured.jpg")
        break

cap.release()
cv2.destroyAllWindows()

# --------------------------
# Safety Check
# --------------------------
if captured is None:
    print("❌ ERROR: No image was captured.")
    exit()

# --------------------------
# Preprocess
# --------------------------
gray = cv2.cvtColor(captured, cv2.COLOR_BGR2GRAY)
resized = cv2.resize(gray, (IMG_SIZE, IMG_SIZE))
img = resized.astype("float32") / 255.0
img = img.reshape(1, IMG_SIZE, IMG_SIZE, 1)

# --------------------------
# Predict
# --------------------------
pred = model.predict(img)[0]
class_idx = np.argmax(pred)
confidence = pred[class_idx]

print("\n==============================")
print("   PREDICTION RESULT")
print("==============================")
print(f"Predicted class : {gesture_classes[class_idx]}")
print(f"Confidence      : {confidence:.4f}")
print("==============================")
```
